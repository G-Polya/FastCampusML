```python
import numpy as np
target = np.random.randint(100)

print(target)
count = 0

while(True):
    trial = int(input('숫자를 맞춰보세요: '))
    if target > trial: # target이 trial보다 크면 큼! 이라고 출력
        count += 1
        print('큼!')
    elif target < trial: # 작으면 작음! 이라고 출력
        count += 1
        print('작음!')
    else:
        count+=1
        print('정답!')
        print(count,'번 만에 정답을 맞추셨습니다')
     
        quit = input('Quit(Y/N): ')
        if quit == 'Y':
            print('게임 종료')
            break
        elif quit == 'N':
            print('계속합니다')
            target = np.random.randintrange(100)
         
            count = 0 
            continue
        
```
