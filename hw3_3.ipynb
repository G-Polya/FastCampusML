```python
import numpy as np
import matplotlib.pyplot as plt


def Distance(a, b):
    return (abs(a[0]- b[0]) / 2) + (abs(a[1] - b[1])/40) + (abs(a[2]-b[2])/5000)



data = np.array([[1,54,3800,1],
                 [1,20,1200,0],
                 [0,45,4200,1],
                 [0,38,4800,1],
                 [1,29,2300,0],
                 [0,30,3500,0],
                 [1,48,2200,1],
                 [0,32,2000,0],
                 [0,25,3300,1]])



kim= np.array([1,24,2500])

# compute distance
distance = list()
for d in data:
    distance.append(Distance(d,kim))
    
distance = np.array(distance)
print(distance)

x = distance
y = data.T[3] # 1,0,1,1,0,0,1,0,1 유권자들의 투표 데이터, 여당(+) : 1, 야당(-) : 0


plt.xlabel("Distance")
plt.ylabel("Vote")

plt.scatter(x,y, s=50)



plt.show()
# y축은 신경쓰지 말고 x축만 보자


# distance-weighted
weight = 1 / distance
weight_sum= np.sum(weight)

y = np.sum((distance / weight_sum) * data.T[3])
if y < 0.5:
    print("Kim(1,24,2500) vote - ")
else:
    print("Kim(1,24,2500) vote + ")
```
